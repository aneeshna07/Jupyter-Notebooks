```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('/content/Project_Dataset_1.csv')
```

```python
data
```

```python
df = data[data['Subgroup'] == 'Total Rape Victims'].drop('Subgroup', axis = 1)
```

```python
df['Unreported_Cases'] = df['Victims_Total'] - df['Rape_Cases_Reported']
df[df['Unreported_Cases'] != 0]
```

```python
kinds = df.columns[3:]
```

## **Forecasting for the Country**

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result=seasonal_decompose(df_yearwise_total['Rape_Cases_Reported'], period=1)
result.plot()
plt.figure(figsize = (200,200))
```

```python
df_diff_1 = df_yearwise_total['Rape_Cases_Reported'] - df_yearwise_total['Rape_Cases_Reported'].shift(1)
df_diff_1.dropna(inplace=True)

plt.figure(figsize=(16,8))
plt.plot(df_diff_1)
```

```python
import statsmodels.api as sm
```

```python
sm.graphics.tsa.plot_acf(df_diff_1, lags=5)
plt.show()
```

```python
sm.graphics.tsa.plot_pacf(df_diff_1, lags=3)
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
adf_result = adfuller(df_diff_1, autolag = 'AIC')
```

```python
def print_adf_results(adf_result):
    print('ADF Statistic: %f' % adf_result[0])
    print('p-value: %f' % adf_result[1])
    print('Critical Values:')
    for key, value in adf_result[4].items():
        print('\t%s : %.3f' % (key, value))
```

```python
print_adf_results(adf_result)
```

```python
A = pm.auto_arima(df_yearwise_total['Rape_Cases_Reported'], start_p=0, start_q=0, d=1, max_p = 3, max_q = 3, seasonal = False, trace = True)
A.fit(df_yearwise_total['Rape_Cases_Reported'])
```

```python
forecast=A.predict(n_periods = 5)
forecast
frame={'Rape_Cases_Projected':forecast, 'Year':[x for x in range(2011,2016)]}
df_forecast=pd.DataFrame(frame)
df_forecast=df_forecast.set_index('Year')
df_forecast['Rape_Cases_Projected'] = [int(x) for x in df_forecast['Rape_Cases_Projected'] ]
```
